```python
import pandas as pd
from pywaffle import Waffle
import matplotlib.pyplot as plt
```

```python
# Charger les données depuis un fichier CSV
file_path = './HP_CHAR_clean.csv'
data = pd.read_csv(file_path)

# Compter le nombre de personnages par maison
house_counts = data['School'].value_counts()

# Créer un graphique Waffle
plt.figure(
    FigureClass=Waffle,
    rows=5,
    columns=20,
    values=house_counts,
    legend={'loc': 'upper left', 'bbox_to_anchor': (1.05, 1)}
)

# Ajouter un titre
plt.title('Répartition des personnages par maison')

# Afficher le graphique
plt.show()
```
