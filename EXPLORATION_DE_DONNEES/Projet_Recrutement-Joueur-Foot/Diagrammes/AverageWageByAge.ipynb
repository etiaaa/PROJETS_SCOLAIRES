```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Charger les données depuis le fichier Excel
# Assure-toi d'adapter le chemin du fichier Excel en fonction de l'emplacement de ton fichier
excel_file_path = '../BDD/data_fifa_clean.xlsx'
df = pd.read_excel(excel_file_path)
```

```python
# Trier les joueurs par popularité dans l'ordre décroissant
df_sorted = df.sort_values('International Reputation', ascending=False)

# Sélectionner un joueur par pays (les 20 premiers)
top_players_by_nationality = df_sorted.groupby('Nationality').head(1).head(20)
```

```python
# Créer le graphique évolutif des salaires en fonction de l'ancienneté avec des intervalles spécifiques
plt.figure(figsize=(12, 6))
sns.lineplot(x='Age', y='Wage', data=df, ci=None, color='purple', marker='o')
plt.title('Évolution Moyenne des Salaires en Fonction de l\'Ancienneté')
plt.xlabel('Âge du Joueur')
plt.ylabel('Salaire Moyen (Wage)')
plt.yticks(range(0, 20001, 500))
plt.tight_layout()
plt.show()
```
