```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # Importer seaborn pour des palettes de couleurs élégantes
```

```python
# Charger les données depuis le fichier Excel
# Assure-toi d'adapter le chemin du fichier Excel en fonction de l'emplacement de ton fichier
excel_file_path = '../BDD/data_fifa_clean.xlsx'
df = pd.read_excel(excel_file_path)
```

```python
# Calculer l'âge moyen par position
age_moyen_par_poste = df.groupby('Position')['Age'].mean().sort_values(ascending=False)
```

```python
# Créer le graphique en boîte
plt.figure(figsize=(12, 10))
ax = sns.boxplot(x='Position', y='Age', data=df, palette='Set2')
ax.set_title('Moyenne d\'âge des joueurs par poste')
ax.set_xlabel('Poste')
ax.set_ylabel('Âge')
plt.yticks(range(int(df['Age'].min()), int(df['Age'].max()) + 1))
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
