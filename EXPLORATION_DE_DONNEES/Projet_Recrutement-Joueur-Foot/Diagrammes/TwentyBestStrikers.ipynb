```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Charger les données depuis le fichier Excel
# Assure-toi d'adapter le chemin du fichier Excel en fonction de l'emplacement de ton fichier
excel_file_path = '../BDD/data_fifa_clean.xlsx'
df = pd.read_excel(excel_file_path)
```

```python
# Supposons que tu aies une DataFrame appelée 'df' contenant les données des joueurs
top_10_buteurs = df.nlargest(20, 'Finishing')  # Utilise la colonne appropriée pour le nombre de buts (Finishing)
```

```python
plt.figure(figsize=(10, 6))
bars = plt.bar(top_10_buteurs['Name'], top_10_buteurs['Finishing'], color='darkseagreen')

# Ajoute le nombre de buts au-dessus de chaque barre
for bar, buts in zip(bars, top_10_buteurs['Finishing']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.2,
             f'({buts})', ha='center', va='bottom')

plt.xlabel('Joueurs')
plt.ylabel('Nombre de buts')
plt.title('Top 10 des Meilleurs Buteurs')
plt.xticks(rotation=45, ha='right')
plt.show()
```
