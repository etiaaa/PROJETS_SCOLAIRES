```python
#Import des modules nécessaires à savoir requests pour créer,utiliser et excécuter nos requêtes et selenium comme recommander. 
import requests
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
```

```python
# Spécifiez le chemin complet vers chromedriver.exe
path_to_chromedriver = './chromedriver-win32/chromedriver.exe'

# Créez une instance de Service en spécifiant le chemin du chromedriver
chrome_service = Service(path_to_chromedriver)

# Créez le navigateur en utilisant l'instance de Service
driver = webdriver.Chrome(service=chrome_service)

# Naviguez vers l'URL souhaitée
driver.get('https://nekketsu.store/')
```

```python
#Création du tableau où seront stocker les données.
tableau = []

#Création d'une boucle qui récupèrera les noms et les produits des vêtements nouveaux. 
i=1 #Initialisation d'une variable i qui sera incrémenter à chaque tour de la boucle afin de modifier les liens et accéder aux différents produits.
for j in range(40000,1000000000): #limitation des produits à afficher 
    try:
        categorie = driver.find_element("xpath",f'/html/body/div{i}/div/div[2]/div[1]/h1').text
        nom = driver.find_element("xpath",f'/html/body/div[4]/div/div[4]/div[1]/div{i}/div/a/div/h2').text
        prix = driver.find_element("xpath",f'/html/body/div[4]/div/div[4]/div[1]/div{i}/div/a/div/span').text
        image_link = driver.find_element("xpath",f'/html/body/div[3]/div/div[4]/div[1]/div{i}/div/div/a/div/div/img').text
        
        print(categorie)
        print(nom)
        print(prix)
        print(image_link)
        tableau.append([categorie, nom, prix, image_link])
    except:
        print("Donnée Manquante")
    i+=1
```

```python
#Affichage du tableau avec toutes les données stocker.   
print(tableau)
```
